3. 卷积分类：
    1）普通卷积：kernel通道数等于特征图通道数；
    2）逐通道卷积：kernel都是单通道的，每个kernel只和特征图的一个通道进行交互（in= out= group）。
    3）逐点卷积：普通卷积，kernel_size=1，作用是通道间交互（每个卷积核卷积完不同通道会进行相加，得到单通道图。）
